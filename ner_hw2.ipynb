```python
import json
from deeppavlov import configs, build_model, train_model

with configs.ner.ner_ontonotes_bert_mult.open(encoding='utf8') as f:
    ner_config = json.load(f)

ner_config['dataset_reader']['data_path'] = './'  # directory with train.txt, valid.txt and test.txt files
ner_config['metadata']['variables']['NER_PATH'] = './'
ner_config['metadata']['download'] = [ner_config['metadata']['download'][-1]]  # do not download the pretrained ontonotes model

ner_model = train_model(ner_config, download=True)
```

```python
ner_model(['Продам хbox360 120g с кинектом и два джойстика'])
```

```python
marked = []

for text in data.text.values[:1000]:
    # BERT имеет лимит на длину текста в 512 слов, возьмем даже еще меньше
    if len(text.split()) > 100:
        continue
    pred = ner_model([text])
    sent, tags = pred[0][0], pred[1][0]
    
    # достанем только тексты с сущностями
    if len(set(tags[0])) > 1:
        marked.append(list(zip(sent,tags)))
```

Few-shot модель демонстрирует неплохую точность и полноту, несмотря на то, что имеет мало размеченных данных. Она нередко допускает ошибки и, например, размечает 'Kaspersky' как 'I-ACC' или ('Wi', 'I-ACC'),('-', 'O'),('Fi', 'O'). Однако хорошо размечает неизвестные, но похожие слова из train, типа ('кабель', 'B-ACC'), ('Uncharted', 'B-GAME'), ('Одни', 'B-GAME'), ('из', 'I-GAME'), ('нас', 'I-GAME'). Однако не является идеальным решением. 
Очевидный вывод заключается в том, что нужно тщательно и аккуратно подбирать данные для обучения, правильно их размечать и использовать объем побольше.
