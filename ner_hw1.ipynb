```python
from yargy import Parser, rule, or_
from yargy.predicates import in_, in_caseless
from yargy.tokenizer import MorphTokenizer
from yargy.pipelines import morph_pipeline, caseless_pipeline
from yargy.interpretation import fact
from IPython.display import display
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)
```

```python
data = pd.read_csv('pristavki.csv', header=None, names=['text'])
```

```python
Accessory = fact(
    'Accessory',
    ['name', 'version']
)
```

```python
gamepads = rule(
    morph_pipeline(['джойстик', 'геймпад', 'джостик', 'пад', 'контроллер', 
                    'джойстики', 'геймпады', 'джостики', 'пады', 'контроллеры',
                   'джойстика', 'геймпада', 'джостика', 'пада', 'контроллера',
                   'gamepad', 'controller']).interpretation(Accessory.name),
    morph_pipeline(['беспроводной', 'проводной', 'беспроводные', 'проводные', 
                    'wireless', 'ps', 'xbox', 'playstation', 'dualshock', 
                    'dualshock 3', 'dualshock 4']).interpretation(Accessory.version).optional())
```

```python
garnitura = rule(
    morph_pipeline(['зарядное устройство', 'гарнитура', 
                    'kinect', 'kinekt', 'kineckt', 'наушники',
                   'уши', 'вертикальный стенд', 'playstation vr', 'адаптер', 'зарядный комплект',
                   'камера', 'микрофон']).interpretation(Accessory.name),
    morph_pipeline(['беспроводной', 'проводной', 'беспроводные', 'проводные', 
                    'wireless', 'ps', 'xbox', 'playstation']).interpretation(Accessory.version).optional())
```

```python
accessories = or_(gamepads, garnitura).interpretation(Accessory)
```

```python
parser = Parser(accessories)
```

```python
matches = []

for sent in data.text[:9000]:
    for match in parser.findall(sent):
        matches.append(match.fact)
```

```python
for m in matches:
    print(m.name, m.version)
```
